## Step 2: Load GPT-2 Model and Tokenizer

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "gpt2"

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Load model
model = AutoModelForCausalLM.from_pretrained(model_name)

print("Model and tokenizer loaded successfully!")
```

## Step 3: Define Python Coding Question Filter

To ensure the model only answers Python coding questions, we implement a simple filtering mechanism.

This function checks if the prompt contains Python-related keywords such as:

- python
- function
- loop
- list
- dictionary
- class
- code

```python
def is_python_question(prompt):

    python_keywords = [
        "python", "function", "loop", "list", "dictionary",
        "class", "def", "code", "variable", "tuple", "array"
    ]

    prompt = prompt.lower()

    for word in python_keywords:
        if word in prompt:
            return True

    return False
```

## Step 4: Generate Response Using GPT-2

If the input is a Python coding question, we pass it to the GPT-2 model to generate a response.

Otherwise, we return a predefined message indicating that the system only answers Python coding questions.

```python
def generate_response(prompt):

    if not is_python_question(prompt):
        return "Sorry, I can only answer Python coding related questions."

    inputs = tokenizer.encode(prompt, return_tensors="pt")

    outputs = model.generate(
        inputs,
        max_length=100,
        num_return_sequences=1,
        no_repeat_ngram_size=2,
        do_sample=True,
        temperature=0.7
    )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return response
```

## Step 5: Test the System

Now we test the model with different prompts.

1. Python related question
2. Non-coding question

The model should only generate answers for Python-related prompts.

```python
# 1. Testing with a Python coding question

prompt1 = "What is a list in python?"

response1 = generate_response(prompt1)

print("Prompt:", prompt1)
print("\nResponse:\n", response1)
```

```python
# 2. Testing with a non-coding question

prompt2 = "What is the capital of France?"

response2 = generate_response(prompt2)

print("Prompt:", prompt2)
print("\nResponse:\n", response2)
```

```python
while True:

    user_prompt = input("Ask a question (type 'exit' to quit): ")

    if user_prompt.lower() == "exit":
        print("Session ended.")
        break

    response = generate_response(user_prompt)

    print("\nResponse:\n", response)
    print("\n" + "-"*50 + "\n")
```
